```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.linspace(-10, 10, 100)
# Calculate y values based on the linear function
y = 2 * x + 3

# Create the plot
plt.plot(x, y)
# Label the axes
plt.xlabel('x')
plt.ylabel('y')
# Title of the plot
plt.title('Plot of the linear function y = 2x + 3')
# Display the plot
plt.show()
```

### Explanation of Solutions

#### Solution 1: Plotting a Linear Function

The solution involves plotting the linear function $$ y = 2x + 3 $$. Here's a breakdown of the code:

- **Import Libraries**: `matplotlib.pyplot` for plotting and `numpy` for numerical operations.
- **Generate x values**: `np.linspace(-10, 10, 100)` generates 100 points between -10 and 10. This function returns evenly spaced numbers over a specified interval.
- **Calculate y values**: `y = 2 * x + 3` computes the y-values based on the linear equation.
- **Plotting**: `plt.plot(x, y)` creates a line plot of x versus y.
- **Labeling and Display**: Labels for the axes are set, a title is added, and `plt.show()` displays the plot.

This code effectively visualizes the linear relationship between x and y, showing how y increases linearly with x.


#### Solution 2: Plotting Multiple Functions

This solution involves plotting two functions, $$ y = x^2 $$ and $$ y = \sqrt{x} $$, on the same graph:

- **Generate x values**: Same as in Solution 1, but only from 0 to 10 because the square root function is not defined for negative x.
- **Calculate y values**: `y1 = x**2` and `y2 = np.sqrt(x)` calculate the y-values for the square and square root functions, respectively.
- **Plotting**: Two `plt.plot()` calls are used, each representing one of the functions. Different line styles and colors distinguish the two functions.
- **Labeling and Legend**: Includes labels, a title, and a legend to identify which line corresponds to which function.

This visualization helps compare the growth rates of a quadratic function and its square root, highlighting their different behaviors.


#### Solution 3: Plotting Trigonometric Functions

This solution plots three trigonometric functions over one period of $$ 2\pi $$:

- **Generate x values**: `np.linspace(0, 2 * np.pi, 100)` generates values from 0 to $$ 2\pi $$, covering one full cycle of trigonometric functions.
- **Calculate y values**: Uses `np.sin(x)`, `np.cos(x)`, and `np.tan(x)` to compute the sine, cosine, and tangent values.
- **Plotting**: Each function is plotted with a different color, and the tangent function's plot limits the y-axis to avoid extreme values due to its asymptotes.
- **Labeling and Legend**: Each function is labeled, and a legend is used for identification.

This plot demonstrates the periodic nature and different phases of the sine, cosine, and tangent functions.


#### Solution 4: Plotting Exponential Growth

This solution plots the exponential function $$ y = e^x $$:

- **Generate x values**: `np.linspace(-5, 5, 100)` creates x-values from -5 to 5.
- **Calculate y values**: `y = np.exp(x)` computes the exponential of x.
- **Plotting and Display**: The plot is created, labeled, and displayed.

This graph shows the rapid increase of the exponential function as x increases, illustrating exponential growth.


#### Solution 5: Plotting Piecewise Functions

This solution involves plotting a piecewise function defined by different conditions:

- **Generate x values**: `np.linspace(-2, 2, 100)` creates x-values from -2 to 2.
- **Define the piecewise function**: `np.piecewise()` is used to define different functional behaviors based on the condition of x (whether x is less than 0 or greater than or equal to 0).
- **Plotting and Display**: The function is plotted, showing a clear distinction between the two pieces at x = 0.

This plot visually represents how piecewise functions can have different expressions over different intervals.
